```python
import nltk
from nltk.corpus import gutenberg
nltk.download('gutenberg')
```

```python
# import data
Alice = nltk.corpus.gutenberg.raw('carroll-alice.txt')
```

```python
nltk.download('punkt')
```

```python
text = nltk.word_tokenize(Alice)
```

```python
from nltk.tokenize import sent_tokenize
```

```python
sentence= sent_tokenize(Alice) 
```

```python
nltk.download('averaged_perceptron_tagger')
```

```python
pos_tags= nltk.pos_tag(sentence[3])
```

```python
pos_tags
```

```python
fd = nltk.FreqDist(pos_tags)
common_tags = fd.most_common(5)
```
