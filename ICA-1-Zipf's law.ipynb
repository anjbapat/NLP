```python
import nltk
from nltk.corpus import gutenberg
nltk.download('gutenberg')
```

```python
nltk.corpus.gutenberg.fileids()
```

```python
# import data
Alice = nltk.corpus.gutenberg.raw('carroll-alice.txt')
```

```python
# step 0: Preprocessing, removing punctuations and lower case the words
import string
# split into words by white space
words = Alice.split()
words = [word.lower() for word in words]
# remove punctuation from each word
table = str.maketrans('', '', string.punctuation)
stripped = [w.translate(table) for w in words]
print(stripped[:100])
```

```python
# step 1: Word count
Total = len(stripped)
Total
```

```python
#step 1: unique words
len(set(stripped))
```

```python
#  step 1: Frequency of occurence
fdist = nltk.FreqDist(stripped)
fdist
```

```python
import pandas as pd
data = pd.DataFrame(list(fdist.items()), columns = ["Word","Frequency"])

```

```python
data = data.sort_values(by='Frequency',ascending=False)
data
```

```python
data= data.rename_axis('Rank')
data
```
